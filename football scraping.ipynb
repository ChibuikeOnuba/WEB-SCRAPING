## WEB SCRAPING

Scraping the fbref.com football website to obtain current epl stats on various team

```python
import requests
```

```python
standings_url = 'https://fbref.com/en/comps/9/Premier-League-Stats'
```

```python
data = requests.get(standings_url)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(data.text)
```

```python
standings_table = soup.select('table.stats_table')[0]
```

```python
table_links = standings_table.find_all('a')
```

```python
tables_link = [l.get('href') for l in table_links]
```

```python
links = [l for l in tables_link if '/squads/' in l]
links = [f'https://fbref.com{l}'for l in links ]
```

```python
links
```

## Getting Arsenal Matches stat

```python
ars_url = links[0]
```

```python
arsenal_htm = requests.get(ars_url)
```

```python
import pandas as pd
ars_table = pd.read_html(arsenal_htm.text, match='Scores & Fixtures')
```

```python
ars_match_table = ars_table[0]
ars_match_table.head()
```

## Getting Arsenal Shooting Stats

```python
ars_soup = BeautifulSoup(arsenal_htm.text)
```

```python
links = ars_soup.find_all('a')
```

```python
links = [l.get('href') for l in links]
```

Here we get the Arsenal shooting link

```python
links = [l for l in links if l and '/all_comps/shooting/' in l]
links
```

```python
shooting_link = links[0]
```

```python
shooting = requests.get(f'https://fbref.com{shooting_link}')
```

```python
shooting_table = pd.read_html(shooting.text, match= 'Shooting')[0]
```

```python
shooting_table.head()
```

```python
shooting_table.columns = shooting_table.columns.droplevel()
```

```python
shooting_table.head(3)
```

```python
ars_match_table.columns
```

```python
shooting_table.columns
```

Next, the ars_match_table is merged with the shooting_table to get a single dataframe

```python
table_data = ars_match_table.merge(shooting_table[['Date','Sh', 'SoT', 'Dist', 'FK',
       'PK', 'PKatt']], on='Date')
```
